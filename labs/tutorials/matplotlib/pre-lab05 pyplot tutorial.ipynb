# pyplot tutorial

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4])
plt.ylabel('some numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 25., 1)

# red dashes, blue squares, and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### plotting with categorical variables

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### controlling line properties

```python
# use keyword args
x=[1,2,3,4]
y=[1,2,3,4]
plt.plot(x, y, linewidth=2.0)
```

```python
line, = plt.plot(x, y, '-') # tuple unpacking
line.set_antialiased(False) # turning off antialiasing
```

```python
x1 = [1,2,3,4]
y1 = [1,2,3,4]
x2 = [-1,-2,-3,-4]
y2 = [-1,-2,-3,-4]

lines = plt.plot(x1, y1, x2, y2)
# use keyword args
plt.setp(lines, color='r', linewidth=2.0)
# or MATLAB style string value pairs
plt.setp(lines, 'color', 'r', 'linewidth', 2.0)
```

```python
lines = plt.plot([1, 2, 3])

# setp can provide a list of settable line properties
plt.setp(lines)
```

### Working with multiple figures and axes

```python
# gca returns current axes
# gcf returns current figure

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1) # figure 1
plt.subplot(211) # figure 1, subplot 1
plt.plot([1, 2, 3])
plt.subplot(212) # figure 1, subplot 2
plt.plot([4, 5, 6]) 

plt.figure(2) # figure 2
plt.plot([4, 5, 6]) # subplot() already made by default

plt.figure(1) # figure 1, subplot 2 is brought forward
plt.subplot(211) # figure 1, subplot 1 brought forward
plt.title('Easy as 1, 2, 3') # figure 1, subplot 1 title changed
```

### workin' w text

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# histogram of data
n, bins, patches = plt.hist(x, 
                            50, 
                            density=1, 
                            facecolor='g', 
                            alpha=0.75)
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
t = plt.xlabel('my data', fontsize=14, color='red')
plt.show()
```

```python
plt.title(r'$\sigma_i=15$') # r = raw string
```

```python
ax = plt.subplot()

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.ylim(-2, 2)
plt.show()
```
